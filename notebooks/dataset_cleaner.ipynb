```python
import pandas as pd
import numpy as np

df = pd.read_csv('../datasets/zomato_data_raw.csv')
df.info()
```

```python
df.head()
```

```python
clean_df = df[['name', 'online_order', 'book_table', 'rate', 'location', 'rest_type', 'cuisines', 'listed_in(type)', 'listed_in(city)']]
clean_df.columns = ['Name', 'Online_Order', 'Book_Table', 'Rating', 'Location', 'Restaurant_Type', 'Cuisines', 'Listed_Type', 'Listed_City']
clean_df.info()
```

```python
clean_df = clean_df[clean_df['Rating'] != 'NEW']
clean_df = clean_df.dropna(subset=['Rating', 'Location', 'Restaurant_Type', 'Cuisines'])
clean_df.info()
```

```python
hyphen_ratings = clean_df['Rating'].str.contains('-', na=False)

clean_df = clean_df[~hyphen_ratings]
clean_df.info()
```

```python
clean_df['Online_Order'] = clean_df['Online_Order'].map({
    "Yes": True,
    "No": False
})

clean_df['Book_Table'] = clean_df['Book_Table'].map({
    "Yes": True,
    "No": False
})

clean_df['Rating'] = clean_df['Rating'].str.strip()
clean_df['Rating'] = clean_df['Rating'].str.split('/5').str[0]
clean_df['Rating'] = clean_df['Rating'].str.strip().astype(float)

clean_df
```

```python
clean_df.to_csv('../datasets/zomato_data_cleaned.csv', index=False)
```
